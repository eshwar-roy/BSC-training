```python
copy_image=image.copy()
gaussian = cv2.GaussianBlur(copy_image,(9,9),0)
median = cv2.medianBlur(copy_image,5)
bilateral_blurr = cv2.bilateralFilter(copy_image,9,75,75)


plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(gaussian )
plt.title('gaussian')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(median)
plt.title('median')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(bilateral_blurr)
plt.title('bilateral')
plt.axis('off')

plt.tight_layout()
```

```python
gray_image_8bit = np.uint8(gray)

# Apply Binary Thresholding
ret, thresh_binary = cv2.threshold(gray_image_8bit, 127, 255, cv2.THRESH_BINARY)

# Apply Inverse Binary Thresholding
ret, thresh_binary_inv = cv2.threshold(gray_image_8bit, 127, 255, cv2.THRESH_BINARY_INV)

# Display both results side-by-side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(thresh_binary, cmap='gray')
plt.title('Binary Threshold')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(thresh_binary_inv, cmap='gray')
plt.title('Inverse Binary Threshold')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Apply Canny edge detection
edges = cv2.Canny(gray_image_8bit, 100, 200)

# Display the edge-detected image
plt.figure(figsize=(8, 6))
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection (Thresholds: 100, 200)')
plt.axis('off')
plt.show()
```

```python
import tensorflow as tf

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu',
        input_shape=(28,28,1)
    ),

    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),

    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(64, activation='relu'),

    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

model.fit(x_train, y_train, epochs=15, batch_size=32)
```
